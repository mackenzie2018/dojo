```python
import numpy as np
```

```python
import random
import string
```

```python
# Generate random number

rand = random.randint(1,50000)
rand
```

```python
# Generate string of random characters

chars = "".join( [random.choice(string.ascii_lowercase) for t in range(rand)] )
```

```python
# Print, just to take a look

chars
```

```python
# Create list of numbers which correspond to the points at which alphabetically ordered substrings end.

end_of_ordered_substrings = [] # initialise

for char in range(0,len(chars)-1): # Iterates up to last item in list
        if chars[char] > chars[char + 1]:
            end_of_ordered_substrings.append(char)

if chars[-1] < chars[-2]: # Last item in list
    end_of_ordered_substrings.append(rand - 1)


end_of_ordered_substrings
```

```python
# Find the size of the alhabetically ordered substrings in the string by:
#    iterating through the above list and finding the difference between item N and item N - 1

size_of_ordered_substrings = [] # initialise

for point in range(len(end_of_ordered_substrings)-1):
    if point == 0: # Account for zero indexing
        size_of_ordered_substrings.append(end_of_ordered_substrings[point] + 1)
    else:
        size_of_ordered_substrings.append( end_of_ordered_substrings[point] - end_of_ordered_substrings[point - 1])
        
size_of_ordered_substrings.append( end_of_ordered_substrings[-1] - end_of_ordered_substrings[-2]) # Last item in list

size_of_ordered_substrings
```

```python
# Find the biggest length of substring, and produce a list of all the locations of the max length substrings
# in the list of points at which alphabetically ordered substrings end

max_substring_length = np.max(size_of_ordered_substrings) # The max value 

max_substring_points = [] # Initialise

for n, size in enumerate(size_of_ordered_substrings):
    if size == max_substring_length:
        max_substring_points.append(n)
        
max_substring_points
```

```python
# Print all substrings of the string which are ordered

longest_substrings = [] # Initialise

for j in max_substring_points: # May be more than one substring tied for longest
    if j == 0: # If the longest ordered substring starts at the start of the string
        longest_substrings.append(chars[0:end_of_ordered_substrings[j] + 1])
    else:
        # Have to add one to the list element range values to account for python indexing
        longest_substrings.append(chars[end_of_ordered_substrings[j - 1] + 1:end_of_ordered_substrings[j] + 1])
```

```python
print("The maximum number of alphabetically ordered characters in the string is {}\n\n".format(max_substring_length))

print("These alphabetically ordered substrings are:\n\n")

longest_substrings
```
